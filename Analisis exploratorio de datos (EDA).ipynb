```python
custom_palette1 = ["#29c5a2", "#3bb3d0", "#70cfe5", "#6c86db","#575cdd", "#1c6cc3","#7b15b8","#d354a0","#ff6f61", "#ff9d5c", "#ffcc67", "#9ed36a"]
plt.figure(figsize=(10, 2))
for i, color in enumerate(custom_palette1):
    plt.plot([i, i+1], [1, 1], color=color, linewidth=20)
plt.axis("off")
plt.title("Paleta de colores personalizada", fontsize=14)
plt.show()
```

## Variables categóricas

```python

dfgenero = df[df['Genero'].isin(['Femenino', 'Masculino'])]
conteo_genero = dfgenero['Genero'].value_counts()
def autopct_format(pct, all_vals):
    total = sum(all_vals)
    val = int(round(pct * total / 100.0))
    return f'{pct:.1f}%\n({val})'
plt.figure(figsize=(8, 6))
plt.pie(
    conteo_genero, 
    labels=conteo_genero.index, 
    autopct=lambda pct: autopct_format(pct, conteo_genero.values), 
    startangle=90, 
    colors=custom_palette1
)
plt.title("Distribución de Género: Femenino y Masculino")
plt.show()
```

```python
idadic = {0: "Ida y vuelta", 1: "Vuelta", 2: "Ida"}
df['Ida/Vuelta'] = df['Ida/Vuelta'].map(idadic)
conteo_ida_vuelta = df['Ida/Vuelta'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(
    x=conteo_ida_vuelta.index, 
    y=conteo_ida_vuelta.values, 
    palette=custom_palette1
)
plt.title("Distribución de Ida/Vuelta")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia")
plt.xticks(rotation=0)
plt.show()
```

# Correlación entre variables

## Matriz de correlación

```python
df_numeric = df[['Dias','Hora Cu-Cu2', 'Hora Cu2-Cu','Satisfaccion', 'Espera Cu-Cu2', 'Espera Cu2-Cu', 'Frecuencia parado']]
correlation_matrix = df_numeric.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral', vmin=-0.1, vmax=0.1)
plt.show()
```
